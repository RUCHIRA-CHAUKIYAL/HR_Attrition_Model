# DATA PREPROCESSING

```python
#dropping unuseful columns


ds.drop(columns = ['EmployeeCount', 'Over18','StandardHours', 'EmployeeNumber'], inplace = True)
```

```python
#new dataframe

ds
```

```python
from sklearn import preprocessing
```

```python
ds['Attrition']=ds['Attrition'].replace({'No':0,'Yes':1})
```

```python
ds['OverTime']=ds['OverTime'].map({'No':0,'Yes':1})
```

```python
ds['Gender']=ds['Gender'].map({'Male':0,'Female':1})
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoding_cols=['BusinessTravel','Department','EducationField','JobRole','MaritalStatus']
label_encoders={}

for column in encoding_cols:
    label_encoders[column]=LabelEncoder()
    ds[column]=label_encoders[column].fit_transform(ds[column])
```

```python
ds.head()
```

```python
data_model=ds
```

```python
data_model
```

# Machine Learning

```python
x=data_model.drop(['Attrition'],axis=1)  #input feature 
y=data_model['Attrition']                #output feature
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

# Feature Selection

```python
from sklearn.ensemble import ExtraTreesRegressor
extractor=ExtraTreesRegressor()
```

```python
extractor.fit(x_train,y_train)
```

```python
x_columns=x_train.columns
feature_rank=pd.DataFrame({'features':x_columns,'importance':extractor.feature_importances_})
```

```python
feature_rank=feature_rank.sort_values('importance',ascending=False)
print(feature_rank)
```

# Model Building

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error
```

```python
def predict(ml_model):
    
    print('Model name is: {}'.format(ml_model))
    model = ml_model.fit(x_train,y_train)
    print('Training score: {}'.format(model.score(x_train,y_train)))
    
    predictions = model.predict(x_test)   #Predictions on Testing data
    
    r2score = r2_score(y_test,predictions)
    print('R2 Score is: {}'.format(r2score))
    
    print('MAE: {}'.format(mean_absolute_error(y_test,predictions)))
    print('MSE: {}'.format(mean_squared_error(y_test,predictions)))
    print('RMSE: {}'.format(np.sqrt(mean_squared_error(y_test,predictions))))
```

```python
predict(LinearRegression())
```

```python
predict(DecisionTreeRegressor())
```

```python
predict(RandomForestRegressor())
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
params={'n_estimators':[100,200,300,400,500],'max_depth':[5,10,15,20],'max_features':['auto','sqrt']}
rf=RandomForestRegressor()
rf_cv=RandomizedSearchCV(rf,params,cv=10,verbose=True,n_jobs=-1)
rf_cv.fit(x_train,y_train)
```

```python
rf_cv.best_params_
```
